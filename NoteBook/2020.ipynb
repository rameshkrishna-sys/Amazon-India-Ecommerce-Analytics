```python
#Q1 datetime format 

year_2020['order_date'] = pd.to_datetime(
    year_2020['order_date'],
    format='mixed',
    dayfirst=True,
    errors='coerce'
)

# Fill invalid dates
median_date = year_2020['order_date'].dropna().median()

year_2020['order_date'] = (
    year_2020['order_date']
    .fillna(median_date)
)

# Standardize format
year_2020['order_date'] = (
    year_2020['order_date']
    .dt.strftime('%Y-%m-%d')
)
```

```python
#Q2 -  column to contain only numeric values in Indian Rupees.

year_2020['original_price_inr'] = year_2020['original_price_inr'].str.replace('₹', '', regex=False)
year_2020['original_price_inr'] = year_2020['original_price_inr'].str.replace('Rs', '', regex=False)
year_2020['original_price_inr'] = year_2020['original_price_inr'].str.replace(',', '', regex=False)
year_2020['original_price_inr'] = year_2020['original_price_inr'].astype(float)

year_2020['original_price_inr'].iloc[16:45]
```

```python
#Q3 - column to contain only numeric values representing the rating.
year_2020['customer_rating'] = year_2020['customer_rating'].astype(str) # Convert to string for processing
year_2020['customer_rating'] = year_2020['customer_rating'].str.split('/').str[0] # Keep only the part before the slash
year_2020['customer_rating'] = year_2020['customer_rating'].str.replace('stars', '', regex=False) # Remove 'stars' and empty strings if present
year_2020['customer_rating'] = year_2020['customer_rating'].str.strip() # Remove any leading/trailing whitespace
year_2020['customer_rating'] = pd.to_numeric(year_2020['customer_rating'], errors='coerce') # Convert to numeric, coercing errors to NaN
```

```python
year_2020['customer_city'].unique()
```

```python
#Q4 - column to contain standardized city names.
city_map = {
    # Mumbai
    'Bombay': 'Mumbai',
    'mumba': 'Mumbai',
    'Mumbai ': 'Mumbai',
    
    # Bengaluru
    'Bengalore': 'Bengaluru',
    'Bengaluru': 'Bengaluru',
    'Banglore': 'Bengaluru',
    'Bangalore': 'Bengaluru',
    'Bangalore ': 'Bengaluru',
    
    # Chennai
    'Madras': 'Chennai',
    'Chennai': 'Chennai',
    'Chennai ': 'Chennai',
    'Chenai': 'Chennai',
    'Chenai ': 'Chennai',
    
    # Kolkata
    'Calcutta': 'Kolkata',
    'Kolkata ': 'Kolkata',
    
    # Delhi
    'New Delhi': 'Delhi',
    'Delhi NCR': 'Delhi'
}


year_2020['customer_city'] = year_2020['customer_city'].replace(city_map)

year_2020['customer_city'] = year_2020['customer_city'].str.title()
year_2020['customer_city'].unique()
```

```python
#Q5 - columns to contain boolean values (True/False).

bool_cols = ['is_prime_member', 'is_prime_eligible', 'is_festival_sale']

bool_map = {
    'True': True,
    'TRUE': True,
    'Yes': True,
    'Y': True,
    '1': True,
    1: True,
    True: True,

    'False': False,
    'FALSE': False,
    'No': False,
    'N': False,
    '0': False,
    0: False,
    False: False
}

for col in bool_cols:
    year_2020[col] = year_2020[col].map(bool_map)
```

```python
year_2020[bool_cols].head()
```

```python
#Q6 - column to contain standardized product categories.

category_map = {
    'ELECTRONICS': 'Electronics',
    'Electronic': 'Electronics',
    'Electronicss': 'Electronics',
    'Electronics & Accessories': 'Electronics'
}

year_2020['category'] = (
    year_2020['category']
    .replace(category_map)
)
```

```python
year_2020['category'].value_counts()
```

```python
#Q7 - column to contain numeric values representing delivery days.
# Replace text values with numeric equivalents
year_2020['delivery_days'] = year_2020['delivery_days'].replace({
    'Same Day': 0,
    'Express': 1,
    '1-2 days': 2
})

# Convert to numeric
year_2020['delivery_days'] = pd.to_numeric(
    year_2020['delivery_days'],
    errors='coerce'
)

# Calculate median from valid values
valid_median = year_2020.loc[
    (year_2020['delivery_days'] >= 0) &
    (year_2020['delivery_days'] <= 30),
    'delivery_days'
].median()

# Replace invalid values
year_2020.loc[
    (year_2020['delivery_days'] < 0) |
    (year_2020['delivery_days'] > 30) |
    (year_2020['delivery_days'].isna()),
    'delivery_days'
] = valid_median

# Convert to integer
year_2020['delivery_days'] = year_2020['delivery_days'].astype(int)
```

```python
year_2020['delivery_days'].value_counts()
```

```python
# Q8: Duplicate Detection & Removal

# Define columns that represent a unique transaction item
dup_cols = ['customer_id', 'product_id', 'order_date','quantity' ,'final_amount_inr', 'payment_method']

# Check number of duplicates
duplicate_count = year_2020.duplicated(subset=dup_cols).sum()
print("Number of duplicate rows:", duplicate_count)

# Remove duplicates
year_2020 = year_2020.drop_duplicates(subset=dup_cols, keep='first')

# Verify
print("Duplicates after removal:", year_2020.duplicated(subset=dup_cols).sum())

```

```python
catalog = pd.read_csv(r"C:\Users\Admin\Documents\amazon\CSV\amazon_india_products_catalog.csv")

catalog = catalog[['product_id', 'base_price_2015']]
```

```python
#Q9 - column to contain the base price of the product in 2020.
year_2020 = year_2020.merge(catalog, on='product_id', how='left') # Merge to get base price
year_2020['price_ratio'] = year_2020['original_price_inr'] / year_2020['base_price_2015'] # Calculate price ratio to identify outliers
mask = (year_2020['price_ratio'] < 0.3) | (year_2020['price_ratio'] > 3) # Define outliers as those with price ratio less than 0.3 or greater than 3


year_2020.loc[mask, 'original_price_inr'] = year_2020['base_price_2015'] # Replace outliers with base price
year_2020.drop(columns=['base_price_2015', 'price_ratio'], inplace=True) # Clean up temporary columns
```
