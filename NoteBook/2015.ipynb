```python
import pandas as pd
import numpy as np

# Replace text values with numeric equivalents
year_2015['delivery_days'] = year_2015['delivery_days'].replace({
    'Same Day': 0,
    'Express': 1,
    '1-2 days': 2
})

# Convert to numeric
year_2015['delivery_days'] = pd.to_numeric(
    year_2015['delivery_days'],
    errors='coerce'
)

# Calculate median from valid values
valid_median = year_2015.loc[
    (year_2015['delivery_days'] >= 0) &
    (year_2015['delivery_days'] <= 30),
    'delivery_days'
].median()

# Replace invalid values
year_2015.loc[
    (year_2015['delivery_days'] < 0) |
    (year_2015['delivery_days'] > 30) |
    (year_2015['delivery_days'].isna()),
    'delivery_days'
] = valid_median

# Convert to integer
year_2015['delivery_days'] = year_2015['delivery_days'].astype(int)
```

```python
category_median = (
    year_2015
    .groupby('category')['original_price_inr']
    .transform('median')
)
```

```python
mask_100x = (
    year_2015['original_price_inr']
    > category_median * 100
)

mask_10x = (
    year_2015['original_price_inr']
    > category_median * 10
)
```

```python
# Fix 100x errors
year_2015.loc[
    mask_100x,
    'original_price_inr'
] = (
    year_2015.loc[
        mask_100x,
        'original_price_inr'
    ] / 100
)

# Fix 10x errors
year_2015.loc[
    (~mask_100x) & mask_10x,
    'original_price_inr'
] = (
    year_2015.loc[
        (~mask_100x) & mask_10x,
        'original_price_inr'
    ] / 10
)
```
